# Pyplot Formatting

## Setting up Environment

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# load dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv")
dataset
```

## Plotting Consecutive Plots using Matplotlib
Plotting data on multiple consecutive figures can be done by calling the corresponding graphing functions and displaying each figure consecutively

```python
# show the first plot (red color)
plt.plot(dataset.work_force, dataset.income, color='r')
plt.show()

# show the second plot
plt.plot(dataset.physicians, dataset.income)
plt.show()
```

### Combining the 2 plots into 1

```python
plt.plot(dataset.work_force, dataset.income, color='r')
plt.plot(dataset.physicians, dataset.income)

plt.show()
```

## Making the Plots Descriptive

### Adding a Title

```python
plt.scatter(dataset.percent_senior, dataset.crime_rate)

# adding a title to the plot
plt.title('Plot of Crime Rate vs Percent Senior')
plt.show()
```

### Adding Labels

```python
plt.scatter(dataset.percent_senior, dataset.crime_rate)

# adding a title to the plot
plt.title('Plot of Crime Rate vs Percent Senior')

# adding the label for the horizontal axis
plt.xlabel('Percent Senior')
# adding the label for the vertical axis
plt.ylabel('Crime Rate')
plt.show()
```

### Adding a legend

```python
# we're simple adding labels to the plots that will be displayed in a small box in the corner
plt.plot(dataset.work_force, dataset.income, color='r', label='work_force')
plt.plot(dataset.physicians, dataset.income, label='physicians')

# adding a legend
plt.legend()
plt.show()
```

## Formatting the plots


### Changing Colours

```python
plt.plot(dataset.physicians, dataset.income, color='blue')    # specify color by name
plt.plot(dataset.work_force, dataset.income, color='c')       # short color code (rgbcmyk)
plt.plot(dataset.hospital_beds, dataset.income, color='0.75') # Grayscale between 0 and 1

#Even any of the below will work:
# plt.plot(dataset.work_force, dataset.income, color='#FFDD44')        # Hex code (RRGGBB from 00 to FF)
# plt.plot(dataset.work_force, dataset.income, color=(1.0, 0.2, 0.3))  # RGB tuple, values 0 to 1
# plt.plot(dataset.work_force, dataset.income, color='chartreuse')     # all HTML color names supported

plt.show()
```

### Changing Line Styles

```python
plt.plot(dataset.work_force, dataset.income, linestyle='dashed')
# plt.plot(dataset.work_force, dataset.income, linestyle='--') # shortcut for dashed line

# Other styles
# plt.plot(dataset.work_force, dataset.income, linestyle='solid') # default style
# plt.plot(dataset.work_force, dataset.income, linestyle='-')     # shortcut for solid line

# plt.plot(dataset.work_force, dataset.income, linestyle='dashdot')
# plt.plot(dataset.work_force, dataset.income, linestyle='-.')  # shortcut for dashdot line

# plt.plot(dataset.work_force, dataset.income, linestyle='dotted')
# plt.plot(dataset.work_force, dataset.income, linestyle=':') # shortcut for dotted line

plt.xlabel('Work Force')
plt.ylabel('Income')
plt.show()
```

### Adding Markers

```python
plt.plot(dataset.work_force, dataset.income, marker='o') # circular markers
# plt.plot(dataset.work_force, dataset.income, marker='x') # crosses markers
# plt.plot(dataset.work_force, dataset.income, marker='^') # triangles markers

plt.xlabel('Work Force')
plt.ylabel('Income')
plt.show()
```

## Changing Plot Size

```python
plt.figure(figsize=(12, 5)) # 12x5 plot

plt.plot(dataset.work_force, dataset.income)
plt.xlabel("Work Force")
plt.ylabel("Income")
plt.show()
```

## Multiple Plots

### Changing Plot Size

```python
fig, ax = plt.subplots()
print(fig, ax)
```

### 1 row and 2 columns - Method 1

```python
plt.subplot(1, 2, 1) # row, column, index
plt.plot(dataset.work_force, dataset.income, "go")
plt.title("Income vs Work Force")

plt.subplot(1, 2, 2) # row, column, index
plt.plot(dataset.hospital_beds, dataset.income, "r^")
plt.title("Income vs Hospital Beds")

plt.suptitle("Sub Plots") # Add a centered title to the figure.
plt.show()
```

### 1 row and 2 columns - Method 2

```python
fig, (ax1, ax2) = plt.subplots(1, 2) #row, column
fig.suptitle("Horizontally stacked subplots")

ax1.plot(dataset.work_force, dataset.income, "go")
ax1.set_title("Income vs Work Force")

ax2.plot(dataset.hospital_beds, dataset.income, "r^")
ax2.set_title("Income vs Hospital Beds")

plt.show()
```

### Overlapping Values

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True) #row, column
fig.suptitle("Horizontally stacked subplots")

ax1.plot(dataset.work_force, dataset.income, "go")
ax1.set_title("Income vs Work Force")

ax2.plot(dataset.hospital_beds, dataset.income, "r^")
ax2.set_title("Income vs Hospital Beds")

plt.show()
```

#### 2 rows and 1 column

```python
plt.subplot(2, 1, 1) # row, column, index
plt.plot(dataset.work_force, dataset.income, "go")

plt.subplot(2, 1, 2) # row, column, index
plt.plot(dataset.hospital_beds, dataset.income, "r^")

plt.suptitle("Sub Plots")
plt.show()
```

### 2 rows and 2 columns


```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6), sharey=True, sharex=True)
ax[0, 0].plot(dataset.work_force, dataset.income, "go") # top-left axes
ax[0, 1].plot(dataset.hospital_beds, dataset.income, "bo") # top-right axes
ax[1, 0].plot(dataset.percent_senior, dataset.income, "yo") # bottom-left axes
ax[1, 1].plot(dataset.physicians, dataset.income, "ro") # bottom-right axes

plt.show()
```
